```python
import random
import pandas as p
[w1,w2,w3,w4,w5,w6]=[0.39825,0.2495,0.51575,0.2935,0.307,0.3035]
value=[]
for i in range(100):
    x1=random.randint(1,10)
    x2=random.randint(1,10)
    x3=random.randint(1,5)
    x4=random.randint(0,100)
    x5=random.randint(1,100)
    x6=random.randint(1,100)
    satisfaction=x1*w1+x2*w2+x3*w3+x4*w4+x5*w5+x6*w6
    cl=0
    if satisfaction>=70:
        cl=2
    elif satisfaction>=40 and satisfaction<70:
        cl=1
    else:
        cl=0
    value.append([x1,x2,x3,x4,x5,x6,satisfaction,cl])
d=p.DataFrame(value,columns=['COMMUNICATION','BEHAVIOUR','TECHNOLOGY','COLLABORATION','ACCURATE DETAILS','SECURITY','SATISFACTION','SATISFIED'])
d.to_csv('hackathon6.csv',index=False)    
 
```

```python
import pandas as pd 
import matplotlib.pyplot as plt

#DATA HANDLING
data=pd.read_csv("hackathon6.csv")
print(data.head())
print(data.describe())
x=data[["COMMUNICATION","BEHAVIOUR","TECHNOLOGY","COLLABORATION","ACCURATE DETAILS","SECURITY","SATISFACTION","SATISFIED"]]
Y=data[["COMMUNICATION","BEHAVIOUR","TECHNOLOGY","COLLABORATION","ACCURATE DETAILS","SECURITY","SATISFACTION","SATISFIED"]]
y=data["SATISFACTION"]
y_=data["SATISFIED"]

#DATA ANALYSIS
plt.scatter(x["COMMUNICATION"],Y["BEHAVIOUR"],color='b')
plt.xlabel("COMMUNICATION")
plt.ylabel("BEHAVIOUR")
plt.show()
```
